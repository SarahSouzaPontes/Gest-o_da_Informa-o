```
import pandas as pd

# Especificar o caminho para o primeiro arquivo Excel no Google Drive
caminho_arquivo1 = '/content/sample_data/WYDEN.xlsx'

# Especificar o caminho para o segundo arquivo Excel no Google Drive
#caminho_arquivo2 = '/content/sample_data/EstacioD.xlsx'

# Ler ambos os arquivos Excel como conjuntos de dados (datasets)
dataset_estacio = pd.read_excel(caminho_arquivo1)
#dataset_estacio2 = pd.read_excel(caminho_arquivo2)

# Unir os conjuntos de dados
#dataset_estacio= pd.concat([dataset_estacio1, dataset_estacio2])

# Verificar as primeiras linhas do conjunto de dados unido para garantir que a união foi feita corretamente
print(dataset_estacio.head())
```

```
# Definindo o dicionário de dados para dataset_galena


dados_galena = {
    "catalog_hide": "Falso",
#catalog_hide
#Verdadeiro
#Falso
#Todos são falsos até que seja sinalizado algum verdadeiro.

    "cursos_horários.id": dataset_estacio['CHAVE'],
    "Chaveamento (external_id)": "",
    "Curso (título do card)": dataset_estacio['NOME_CURSO'].apply(lambda x: x.title()),
    "Informação Complementar (Subtítulo do card)": "",
    "Descrição do curso (HTML)": "",
    "Duração do Curso": dataset_estacio['QTD_MIN_PERIODO'],
    "Duração-tipo": "Hora",
    "É mensalidade?": "Sim",
#É mensalidade?
#Sim
#Não
#Cursos da estácio são sempre mensalidade. Ou seja, sempre "sim".

    "Parcelas": dataset_estacio['QTD_MIN_PERIODO'],
    "Valor da mensalidade/Valor total do curso": dataset_estacio['PRECO_CHEIO'].map(lambda x: f"R$ {x:.2f}"),
    "Desconto exclusivo Edupass (%)": dataset_estacio['DESCONTO_PARCEIRO_ONLINE'].map(lambda x: f"{x:.2f}%"),
    "Categoria": dataset_estacio['GRAU_ACADEMICO'].apply(lambda x: x.title()),

#Categoria
#Cuta Duração
#Cursos Técnicos
#Desenvolvimento Pessoal
#Ensino Básico
#Graduação
#Idiomas
#Pé-Vetsibular
#Pós Graduação
#Para crianças e adolescentes
#Tech & Digital
# Etácio - Sempre Graduação.

    "Categoria específica": dataset_estacio['GRAU_ACADEMICO'].apply(lambda x: x.title()),
#Categoria Específicas
#Bacharelado
#Doutorado
#Mestrado
#Especialização
#Licenciatura
#MBA
#Tecnologia
# Na estácio Categoria Específica é = a Grau Acadêmico. Manter primeira maíuscula e reduzir a caixa alta das demais letras para a versão final

    "Tipo": dataset_estacio['GRAU_ACADEMICO'].apply(lambda x: x.title()),

#Tipo
#Livre
#Pós Graduação
#Graduação
#Técnico
#Idiomas
#Básico
#Na Estácio será sempre igual a Coluna da "Categoria" do padrão da Galena.

    "Campus/Polo/Unidade/Online": dataset_estacio['NOME_CAMPUS'].apply(lambda x: x.title()),
    "Endereço": dataset_estacio['ENDEREÇO POLO'].apply(lambda x: x.title() if isinstance(x, str) else x),
    "Bairro": "",
    "Cidade": "",
    "UF": dataset_estacio['UF'],
    "Modalidade": dataset_estacio["MODALIDADE"].apply(lambda x: x.title()),
    "Turno": dataset_estacio["NOME_TURNO"].apply(lambda x: x.title()),
#Turno
#Manhã
#Tarde
#Noite
#Flexível
#Integral
#Correspondência da coluna "TURNO", porém deixar a primeira maiúscila e demais caixa baixa.
#Quando tiver mais de um turno na mesma coluna pode substituir por "Flexível".
#Quando o Turno apresentar o valor "Virtual" Substituir por "Flexével"

    "FLAG_BOLSA_REGRESSIVA" : dataset_estacio["FLAG_BOLSA_REGRESSIVA"],
    "BOLSA_REGRESSIVA_PARCEIRO" :  dataset_estacio['BOLSA_REGRESSIVA_PARCEIRO'].map(lambda x: '{:.2%}'.format(x)),
    "NOME_IES" : dataset_estacio["NOME_IES"]

}

# Criar o dataset_galena a partir do dicionário de dados
dataset_galena = pd.DataFrame(dados_galena)
```

```
# Obter os valores únicos da coluna 'Turno'
turnos_unicos = dataset_galena['Turno'].unique()
turnos_unicos
dataset_galena['Turno'].replace(['Manha/Tarde', 'Tarde/Noite', 'Integral', 'Virtual'], 'Flexivel', inplace=True)
```

```
dataset_galena['Turno'] = dataset_galena['Turno'].replace('Manha', 'Manhã', regex=True)
```

```
# Verificar se 'FLAG_BOLSA_REGRESSIVA' é igual a 'SIM'
#LAG_BOLSA_REGRESSIVA - 'Não' / PREÇO_CHEIO / DESCONTO_PARCEIRO_ONLINE
#FLAG_BOLSA_REGRESSIVA - 'SIM' / PREÇO_CHEIO / DESCONTO_PARCEIRO_ONLINE - Informação Complementar (Subtítulo do card) : "Desconto de $BOLSA_REGRESSIVA_PARCEIRO$ a partir do segundo semestre".

condicao = dataset_galena['FLAG_BOLSA_REGRESSIVA'] == 'SIM'

# Preencher 'Informação Complementar (Subtítulo do card)' com a string desejada onde a condição é verdadeira
dataset_galena.loc[condicao, 'Informação Complementar (Subtítulo do card)'] = 'Desconto de ' + dataset_galena.loc[condicao, 'BOLSA_REGRESSIVA_PARCEIRO'] + ' a partir do segundo semestre'
```

```
dataset_galena = dataset_galena.drop(columns=['FLAG_BOLSA_REGRESSIVA', 'BOLSA_REGRESSIVA_PARCEIRO'])
```

```
valores_distintos = dataset_estacio['NOME_IES'].nunique()
print("Número de valores distintos em 'NOME_IES':", valores_distintos)
```

```
# Carregar o dataset_galena
# Verificar os valores distintos na coluna 'NOME_IES'
valores_distintos = dataset_galena['NOME_IES'].unique()

```

```
# Criar um dicionário para armazenar os subconjuntos
subconjuntos = {}
```

```

# Dividir o dataset_galena de acordo com os valores distintos de 'NOME_IES'
for valor in valores_distintos:
    subconjuntos[valor] = dataset_galena[dataset_galena['NOME_IES'] == valor]
```

```

# Agora 'subconjuntos' é um dicionário onde as chaves são os valores distintos de 'NOME_IES' e os valores são os dataframes correspondentes

# Salvar os subconjuntos em arquivos separados (opcional)
for valor, subconjunto in subconjuntos.items():
    nome_arquivo = f"dataset_galena_{valor}.xlsx"  # Nome do arquivo com o valor de 'NOME_IES' no nome
    subconjunto.to_csv(nome_arquivo, index=False)
    print(f"Subconjunto com 'NOME_IES' igual a '{valor}' salvo como '{nome_arquivo}'.")
```
